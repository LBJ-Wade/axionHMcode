```python
def func_k_nl(spec_dic):

    def func_root_sigma(r):
        return variance.func_sigma_r(r, spec_dic['k'], spec_dic['cold']) - 1
    
    r_nl = optimize.brentq(func_root_sigma, 1e-3, 1e3)
    
    return np.pi / r_nl

k_nl_arr = []

for power_spec_dic in power_spec_interp_dic_tabel:
    k_nl_arr.append(func_k_nl(power_spec_dic))
```

```python
#### half mode scales ####

def func_transfer_ratio(transfer_dic, transfer_dic_LCDM, cosmo_dic, cosmo_dic_LCDM):
    # be carefull transfer_dic['k'] is in units of h/Mpc and the tranfer_function in (Mpc)^2
    transfer_total_matter_LCDM_func = interpolate.interp1d(transfer_dic_LCDM['k'], (cosmo_dic_LCDM['Omega_b_0']*transfer_dic_LCDM['transfer_baryon'] + cosmo_dic_LCDM['Omega_d_0']*transfer_dic_LCDM['transfer_CDM']) \
                                                           /cosmo_dic_LCDM['Omega_db_0'], kind='cubic')
 
    return transfer_dic['k'] , (cosmo_dic['Omega_b_0']*transfer_dic['transfer_baryon'] + cosmo_dic['Omega_d_0']*transfer_dic['transfer_CDM']) \
                                                           /cosmo_dic['Omega_db_0'] / transfer_total_matter_LCDM_func(transfer_dic['k']) #transfer_dic_LCDM['transfer_total']

def func_half_mode(cosmo, cosmos_LCM, transfer_dic, transfer_dic_LCDM):
    # be carefull transfer_dic['k'] is in units of h/Mpc and the tranfer_function in (Mpc)^2
    
    k_in_func, transfer_ratio_in_func = func_transfer_ratio(transfer_dic, transfer_dic_LCDM, cosmos, cosmos_LCM) # k in h/Mpc

    transfer_ratio_half_value =  transfer_ratio_in_func[0] - 0.5 * (transfer_ratio_in_func[0] - transfer_ratio_in_func[-5])
    
    func_ratio_of_k = interpolate.interp1d(k_in_func, transfer_ratio_in_func, kind='cubic')
    
    def func_root(k):
        return func_ratio_of_k(k) - transfer_ratio_half_value
        
    half_mode_value = optimize.brentq(func_root, transfer_dic['k'][0], transfer_dic['k'][-1]) # in units of h/Mpc
    return half_mode_value #, transfer_ratio_half_value


transfer_dic_LCDM = lin_power_spectrum.load_transfer_from_file('paramfiles/cosmos_LCDM_transfer_out.dat')
transfer_dic_tabel = []
i = 0
for dic in cosmos_tabel:
    transfer = lin_power_spectrum.load_transfer_from_file('paramfiles/cosmos_{}_transfer_out.dat'.format(i))
    transfer_dic_tabel.append(transfer)
    i = i+1

k_half_model = []
i = 0
for transfer in transfer_dic_tabel:
    k_half_model.append(func_half_mode(cosmos_tabel[i], cosmos_LCDM, transfer, transfer_dic_LCDM))
    i = i+1
```

# Halo Mass Relation

```python
M_cut_arr = []
i = 3
for power_spec_dic in power_spec_interp_dic_tabel[3:]:

    M_cut_arr.append(axion_cutoff_mass.func_cut_mass_axion_halo(cosmos_tabel[i], power_spec_dic))
    i = i+1

def func_smoothing(M, M_cut):
    """
    smoothing for the cut-off of the neutrino halo mass relation
    want smooth transition to zero
    """
    c = 3
    a = M_cut #- c * M_c
    b = M_cut + c * M_cut
    def func_smoothing_intervall(x):
        return np.exp(-1/x) / ( np.exp(-1/x) + np.exp(-1/(1-x)) )

    if M < a:
        return 1e-40
    elif M > b:
        return 1
    else:
        return func_smoothing_intervall((M-a)/(b-a))+1e-20
    
    
def cut_sharp(M, M_cut):
    if M < M_cut:
        return 0.
    else:
        return M
    
    
M_ax_tabel = []
M_arr_cutoff = np.logspace(4, 17, 100)
i = 0
for dic in cosmos_tabel[3:]:
    M_ax = []
    for m in M_arr_cutoff:
        M_ax.append(dic['Omega_ax_0']/dic['Omega_db_0'] * m * func_smoothing(m, M_cut_arr[i]))
    
    M_ax_tabel.append(M_ax)
    i = i+1
        
```
