```python
power_spec_dic_ax = lin_power_spectrum.func_power_spec_dic('paramfiles/cosmos_transfer_out.dat', cosmos)
power_spec_dic_LCDM = lin_power_spectrum.func_power_spec_dic('paramfiles/cosmos_LCDM_transfer_out.dat', cosmos_LCDM)

power_spec_interp_dic_ax = lin_power_spectrum.func_power_spec_interp_dic(power_spec_dic_ax, cosmos)
power_spec_interp_dic_LCDM = lin_power_spectrum.func_power_spec_interp_dic(power_spec_dic_LCDM, cosmos_LCDM)
```

```python
M_arr = np.logspace(cosmos['M_min'], cosmos['M_max'], 100)
axion_params = axion_params.func_axion_param_dic(M_arr, cosmos, power_spec_interp_dic_ax, eta_given=False)
```

```python
hmcode_params = HMcode_params.HMCode_param_dic(cosmos, power_spec_interp_dic_ax['k'], power_spec_interp_dic_ax['cold'])
```

```python
import importlib
importlib.reload(PS_nonlin_axion)
importlib.reload(PS_nonlin_cold)
PS_matter_nonlin = PS_nonlin_axion.func_full_halo_model_ax_sophie(M_arr, power_spec_dic_ax, power_spec_interp_dic_ax, cosmos, hmcode_params, axion_params)

print(time.time() -start)
```

```python
PS_cold = PS_nonlin_cold.func_non_lin_PS_matter(M_arr, power_spec_dic_ax['k'], power_spec_dic_ax['cold'], power_spec_interp_dic_ax['k'], power_spec_interp_dic_ax['cold'], 
                                                              cosmos, hmcode_params, cosmos['Omega_db_0'], cosmos['Omega_db_0'], one_halo_damping = True)
```

```python
hmcode_params_LCDM = HMcode_params.HMCode_param_dic(cosmos_LCDM, power_spec_interp_dic_LCDM['k'], power_spec_interp_dic_LCDM['cold'])
```

```python
PS_LCDM_matter_nonlin = PS_nonlin_cold.func_non_lin_PS_matter(M_arr, power_spec_dic_LCDM['k'], power_spec_dic_LCDM['power_total'], power_spec_interp_dic_LCDM['k'], power_spec_interp_dic_LCDM['cold'], 
                                                              cosmos_LCDM, hmcode_params_LCDM, cosmos_LCDM['Omega_m_0'], cosmos_LCDM['Omega_db_0'], one_halo_damping = True)
```

```python
plt.semilogx(power_spec_dic_ax['k'], PS_matter_nonlin[0]/PS_LCDM_matter_nonlin[0], label='total matter')
plt.axhline(y = 1, linestyle=':', color='black')

k_PS = np.loadtxt("datafile.txt")

plt.semilogx(power_spec_dic_ax['k'], k_PS[:, 1]/PS_LCDM_matter_nonlin[0], label='cold matter')
plt.legend()
plt.show()
```
